```python
import matplotlib.pyplot as plt
import numpy as np
from math import log, floor, ceil
```

```python
def q1(n):
    line3 = 0
    line5 = 0
    count = 0
    j = 1
    while j < n:
        line3 += 1
        j = j * 2
        for k in range(1, j):
            line5 += 1
            count = count + 1
    return {"n": n, "line3": line3, "line5": line5}


def line3(n):
    return q1(n)['line3']


def line5(n):
    return q1(n)['line5']
```

```python
[(n, line5(n)) for n in range(1, 100)]
```

```python
from math import log2

def line5_algo(n):
    total = 0
    for j in range(0,ceil(log2(n))):
        total += (2**j)*2 - 1
    return total
```

```python
for n in range(1, 10000):
    expected = line5(n)
    got = line5_algo(n)
    if expected != got:
        print(f"n={n} expected {expected} but got {got}")
```
